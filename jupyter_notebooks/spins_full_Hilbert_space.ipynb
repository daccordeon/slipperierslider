## Basic checks

Eigenstates

```python
# eigenstates of Jz
M = 3
(
    all(
        Jz(M) @ Jz_up_state(M, return_ket=True)
        == M / 2 * Jz_up_state(M, return_ket=True)
    )
    & all(
        Jz(M) @ Jz_down_state(M, return_ket=True)
        == -M / 2 * Jz_down_state(M, return_ket=True)
    )
)
```

```python
# eigenstates of Jx
M = 3
all(Jx(M) @ Jx_up_state(M, return_ket=True) == M / 2 * Jx_up_state(M, return_ket=True))
```

Commutators

```python
M = 5
(
    np.all(commutator(Jx(M), Jy(M)) == 1j * Jz(M))
    & np.all(commutator(Jy(M), Jz(M)) == 1j * Jx(M))
    & np.all(commutator(Jz(M), Jx(M)) == 1j * Jy(M))
)
```

Dicke states

```python
M = 5

fig, axs = plt.subplots(M + 1, 1, sharex=True, figsize=(8, 6))
for j, ax in enumerate(axs):
    ket = symmeterised_Jz_eigenstate(M, -M / 2 + j, return_ket=True)
    ax.stem(Jz_probabilities(M, ket))
    ax.set(xlim=(-0.5, 2**M - 0.5), ylim=(0, 1))
    ax.text(0.05, 0.5, rf"$\hat{{J}}_z = {-M/2 + j}$", transform=ax.transAxes)
ax.set_xlabel(r"index in reverse lexicographic order of $\hat \sigma_z$ basis")
fig.text(0.05, 0.5, r"probability", va="center", rotation="vertical")
plt.show()
```

## Evolution under TAT over time

M = 5

```python
M = 5
ket0 = Jz_down_state(
    M, return_ket=True
)  # such that when TAT'd it should be analogous to the squeezed vacuum state
data = list(
    symmeterised_Jz_probabilities(M, ket0)
)  # TAT should preserve symmetry such that this is sufficient, TODO: Check
χs = np.linspace(0, np.pi, 100)
for χ in tqdm(χs[1:]):
    ket1 = TAT(M, ket0, χ=χ, truncation=None)
    data1 = symmeterised_Jz_probabilities(M, ket1)
    data.append(data1[1])
data = np.array(data)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
colors = plt.cm.coolwarm(np.linspace(0, 1, M + 1))
for j in reversed(range(len(data[0]))):
    label = data[0, j] if 0 not in data[0] else int(data[0, j])
    ax.plot(χs, data[1:, j], label=label, color=colors[j])
ax.set(
    xlim=(χs.min(), χs.max()),
    ylim=(0, 1),
    xlabel=r"two-axis twisting parameter, $\chi$",
    ylabel=r"probability",
)
ax.legend(title="$\hat J_z$ sym.")
fig.savefig(f"../figures/qubits/evolution_under_TAT_M={M}.pdf", bbox_inches="tight")
plt.show()
```
